```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("diamonds.csv")
```

```python
data
```

```python
sns.set(style="darkgrid")
sns.boxplot(x="cut",y="clarity",data=data)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.boxplot(x="cut",y="clarity",hue="color",data=data)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x="depth",y="price",hue="color",style="color",data=data)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.boxplot(x="cut",y="price",data=data)
plt.show()
```

```python
sns.pairplot(data,hue="color")
plt.show()
```

```python
sns.lineplot(x="cut",y="depth",data=data)
plt.show()
```

```python
sns.lineplot(x="cut",y="depth",hue="color",data=data)
plt.show()
```

```python
sns.lineplot(x="cut",y="depth",hue="color",style="color",data=data)
plt.show()
```

```python
sns.barplot(x="cut",y="price",data=data)
plt.show()
```

```python
sns.barplot(x="cut",y="price",hue="color",data=data)
plt.show()
```

```python
sns.scatterplot(x="depth",y="price",data=data)
plt.show()
```

```python
sns.jointplot(x="cut",y="clarity",data=data)
plt.show()
```

```python
sns.jointplot(x="depth",y="price",hue="color",data=data)
plt.show()
```

```python
sns.barplot(x="cut",y="price",color="red",data=data)
plt.show()
```

```python
sns.barplot(x="cut",y="clarity",color="purple",data=data)
plt.show()
```

```python
sns.jointplot(x="carat",y="table",hue="color",data=data)
plt.show()
```
